```python
import os
import numpy as np

figdir = '../../figure/introduction'

from matplotlib import pyplot as plt

os.makedirs(figdir, exist_ok=True)
```

```python
N = 10
M = 100

a = 532e-9
k = np.pi / a

V_0 = -10

n = np.arange(N+2)
l = N*a*np.linspace(0, 1, 200)

x = a * np.random.randint(1, N, M)
y = a * np.random.randint(1, N, M)

V_lat = V_0*np.cos(k*l)**2
```

```python
fig = plt.figure(figsize=(8, 8))

ax1 = plt.subplot2grid((5, 5), (0, 1), colspan=4, rowspan=4)
ax1.scatter(x / 1e-6, y / 1e-6)
ax1.set_xlabel('x ($a$)')
ax1.set_ylabel('y ($a$)')
ax1.set_xticklabels(n)
ax1.set_yticklabels(n)
ax1.set_xticks(a * n / 1e-6)
ax1.set_yticks(a * n / 1e-6)
ax1.grid(color='gray', linestyle='dotted')
ax1.xaxis.tick_top()
ax1.xaxis.set_label_position('top')
ax1.yaxis.tick_right()
ax1.yaxis.set_label_position('right')

ax2a = plt.subplot2grid((5, 5), (0, 0), colspan=1, rowspan=4, sharey=ax1)
ax2a.plot(V_lat, l / 1e-6)
ax2a.set_xlabel('E ($E_r$)')
ax2a.set_ylabel('y ($\mu m$)')
ax2a.xaxis.tick_top()
ax2a.xaxis.set_label_position('top')

ax2b = plt.subplot2grid((5, 5), (4, 1), colspan=4, rowspan=1, sharex=ax1)
ax2b.plot(l / 1e-6, V_lat)
ax2b.set_ylabel('E ($E_r$)')
ax2b.set_xlabel('x ($\mu m$)')
ax2b.yaxis.tick_right()
ax2b.yaxis.set_label_position('right')

plt.tight_layout()
plt.savefig(os.path.join(figdir, 'lattice-simple.pdf'))
plt.show()
```

```python
V_pot = V_lat.copy()
V_pot[59:139] += 5 * V_0

x_pot = a * np.concatenate([np.random.randint(1, N, M // 5), np.random.randint(3, 7, 50)])
y_pot = a * np.concatenate([np.random.randint(1, N, M // 5), np.random.randint(3, 7, 50)])
```

```python
import matplotlib.patches as patches

fig = plt.figure(figsize=(8, 8))

ax1 = plt.subplot2grid((5, 5), (0, 1), colspan=4, rowspan=4)
ax1.scatter(x_pot / 1e-6, y_pot / 1e-6)
ax1.set_xticks(a * n / 1e-6)
ax1.set_yticks(a * n / 1e-6)
ax1.set_xticklabels(n)
ax1.set_yticklabels(n)
ax1.set_xlabel('x ($a$)')
ax1.set_ylabel('y ($a$)')
ax1.grid(color='gray', linestyle='dotted')
ax1.xaxis.tick_top()
ax1.xaxis.set_label_position('top')
ax1.yaxis.tick_right()
ax1.yaxis.set_label_position('right')
ax1.add_patch(patches.Rectangle((1.75, 1.75), 1.8, 1.8, color='g', alpha=.4))

ax2a = plt.subplot2grid((5, 5), (0, 0), colspan=1, rowspan=4, sharey=ax1)
ax2a.plot(V_pot, l / 1e-6)
ax2a.set_xlabel('E ($E_r$)')
ax2a.set_ylabel('y ($\mu m$)')
ax2a.xaxis.tick_top()
ax2a.xaxis.set_label_position('top')

ax2b = plt.subplot2grid((5, 5), (4, 1), colspan=4, rowspan=1, sharex=ax1)
ax2b.plot(l / 1e-6, V_pot)
ax2b.set_ylabel('E ($E_r$)')
ax2b.set_xlabel('x ($\mu m$)')
ax2b.yaxis.tick_right()
ax2b.yaxis.set_label_position('right')

plt.tight_layout()
plt.savefig(os.path.join(figdir, 'lattice-pot.pdf'))
plt.show()
```
