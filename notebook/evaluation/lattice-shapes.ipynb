```python
import os
import numpy as np

figdir = '../../figure/introduction'

from matplotlib import pyplot as plt

os.makedirs(figdir, exist_ok=True)
```

```python
N = 11
M = 100

X = np.arange(-np.floor(N / 2), +np.ceil(N / 2))

x = np.random.randint(X.min() + 1, X.max(), M)
y = np.random.randint(X.min() + 1, X.max(), M)

Z = np.linspace(X.min(), X.max(), 200)
```

```python
fig = plt.figure(figsize=(8, 8))

ax1 = plt.subplot2grid((5, 5), (0, 1), colspan=4, rowspan=4)
ax1.scatter(x, y)
ax1.set_xlabel('x (a)', )
ax1.set_ylabel('y (a)')
ax1.grid(color='gray', linestyle='dotted')
ax1.set_xticks(X)
ax1.set_yticks(X)
ax1.xaxis.tick_top()
ax1.xaxis.set_label_position('top')
ax1.yaxis.tick_right()
ax1.yaxis.set_label_position('right')

ax2a = plt.subplot2grid((5, 5), (0, 0), colspan=1, rowspan=4)
ax2a.plot(np.sin(2*np.pi*np.linspace(-2.3, 2.3, 200))**2, Z)
ax2a.set_xlabel('E')
ax2a.set_ylabel('y (a)')
ax2a.xaxis.tick_top()
ax2a.xaxis.set_label_position('top')

ax2b = plt.subplot2grid((5, 5), (4, 1), colspan=4, rowspan=1)
ax2b.plot(Z, np.sin(2*np.pi*np.linspace(-2.3, 2.3, 200))**2)
ax2b.set_ylabel('E')
ax2b.set_xlabel('x (a)')
ax2b.yaxis.tick_right()
ax2b.yaxis.set_label_position('right')

plt.tight_layout()
plt.savefig(os.path.join(figdir, 'lattice-simple.pdf'))
plt.show()
```
