```python
import os
import numpy as np

imgdir = '../../media/image'
figdir = '../../figure/intensity/projection'

from PIL import Image
from matplotlib import pyplot as plt

os.makedirs(figdir, exist_ok=True)
```

```python
# pixel size in meter
psize = 5.86e-6
```

```python
lmu = Image.open(os.path.join(imgdir, 'lmu.tif'))
lmu
```

```python
bodo = Image.open(os.path.join(imgdir, 'bodo.tif'))
bodo
```

```python
img = np.array(bodo)
img.shape
```

```python
plt.figure(figsize=(10, 5))
plt.xlabel('$x$ (\si{\milli\meter})')
plt.ylabel('$y$ (\si{\milli\meter})')
plt.imshow(img, extent=[0, psize * img.shape[1] / 1e-3, 0, psize * img.shape[0] / 1e-3])

plt.subplots_adjust(top=.98, bottom=.12)
plt.savefig(os.path.join(figdir, 'bodo.pdf'))
plt.savefig(os.path.join(figdir, 'bodo.pgf'))
plt.show()
```
