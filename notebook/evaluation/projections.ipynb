```python
import os
import numpy as np

imgdir = '../../media/image'
figdir = '../../figure/intensity/projection'

from PIL import Image
from matplotlib import pyplot as plt

os.makedirs(figdir, exist_ok=True)
```

```python
# pixel size in micro meter
psize = 5.86
```

```python
lmu = Image.open(os.path.join(imgdir, 'lmu.tif'))
lmu
```

```python
bodo = Image.open(os.path.join(imgdir, 'bodo.tif'))
bodo
```

```python
img = np.array(bodo)
img.shape
```

```python
plt.figure(figsize=(9, 9))
plt.title('Text Projection via AODs')
plt.xlabel(r'\textit{x} (\si{\micro\meter})')
plt.ylabel(r'\textit{y} (\si{\micro\meter})')
plt.imshow(img, extent=[0, psize * img.shape[1], 0, psize * img.shape[0]])
plt.savefig(os.path.join(figdir, 'bodo.pdf'))
plt.show()
```
