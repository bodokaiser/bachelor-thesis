```python
import os
import numpy as np

imgdir = '../../media/image'
figdir = '../../figure/intensity/projection'

from PIL import Image
from matplotlib import pyplot as plt

os.makedirs(figdir, exist_ok=True)
```

```python
# pixel size in meter
psize = 5.86e-6

# magnification in setup
M = 500 / 4.5
```

```python
lmu = Image.open(os.path.join(imgdir, 'lmu.tif'))
lmu
```

```python
bodo = Image.open(os.path.join(imgdir, 'bodo.tif'))
bodo
```

```python
img = np.array(bodo)
img.shape
```

```python
plt.figure(figsize=(12, 5))
plt.xlabel('$x$ (\si{\micro\meter})')
plt.ylabel('$y$ (\si{\micro\meter})')
im = plt.imshow(img / img.max(), extent=[0, psize * img.shape[1] / M / 1e-6, 0, psize * img.shape[0] / M / 1e-6])

plt.subplots_adjust(top=.96, bottom=.14, hspace=.05, wspace=.05, right=.83)
plt.colorbar(im, label=r'Intensity (\si{\percent})', cax=plt.axes([0.75, 0.14, 0.02, .82]))
plt.savefig(os.path.join(figdir, 'bodo.pdf'))
plt.savefig(os.path.join(figdir, 'bodo.pgf'))
plt.show()
```
