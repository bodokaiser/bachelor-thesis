# Reflection

In this notebook we evaluate the reflection dataset obtained from the network analyzer connected with the acousto-optic deflectors.

```python
import numpy as np
import pandas as pd

from matplotlib import pyplot as plt
```

```python
def read(filename):
    return pd.read_csv(filename, skiprows=3, header=None, usecols=[0, 1], names=['frequency', 'reflection'])
```

```python
h = read('../dataset/reflection/horizontal.csv')
v = read('../dataset/reflection/vertical.csv')
```

```python
plt.title('reflection @10dBm output power')
plt.xlabel('Frequency in Hz')
plt.ylabel('Reflection in dB')
plt.plot(h['frequency'], h['reflection'], label='horizontal AOD')
plt.plot(v['frequency'], v['reflection'], label='vertical AOD')
plt.show()
```
