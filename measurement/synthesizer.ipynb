# Synthesizer

```python
import os
import control
import numpy as np
import pandas as pd

from matplotlib import pyplot as plt

scope = control.MSOX6004A('172.22.22.30')
fngen = control.FG33250A2('172.22.22.36')

scope.holla(), fngen.holla()
```

```python
duration = 26.84e-3

control.aod_h(frequency=[90e6, 110e6], amplitude=np.ones(1024), duration=duration)
```

```python
data = []

for delay in np.linspace(0, duration, 260):
    fngen.width(delay)
    
    data.append(scope.capture())
```

```python
plt.specgram(data[0].voltage.values, NFFT=1024, Fs=1/(data[0].time[1]-data[0].time[0]), noverlap=900)
plt.xlabel('time in s')
plt.ylabel('frequency in Hz')
plt.show()
```

```python
plt.specgram(data[200].voltage.values, NFFT=1024, Fs=1/(data[200].time[1]-data[200].time[0]), noverlap=900)
plt.xlabel('time in s')
plt.ylabel('frequency in Hz')
plt.show()
```
